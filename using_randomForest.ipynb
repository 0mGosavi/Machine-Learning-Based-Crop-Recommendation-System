```python
import pandas as pd                                   # to manipulate dataset
from sklearn.model_selection import train_test_split  # to train x and y data
from sklearn.ensemble import RandomForestClassifier   # using Random forest classifier library
from sklearn.metrics import classification_report, accuracy_score #for printing accuracy
import matplotlib.pyplot as plt # for plotting graphs
%matplotlib inline
```

```python
data = pd.read_csv('Crop_recommendation.csv')   # importing our data set
```

```python
X = data[['temperature', 'ph', 'humidity']]     # Storing value of temperature, humidity and ph value data to X variable
Y = data['label']                               # Storing Label names (Plant names) in Y variable
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X['temperature'], Y, color='r', label='Temperature')  # red color for Temperature
plt.scatter(X['ph'], Y, color='g', label='pH')                    # green color for pH value
plt.scatter(X['humidity'], Y, color='b', label='Humidity')        # blue color for Humidity
plt.xlabel('Values')      # name of x axis
plt.ylabel('Plant Names') # name of Y axis
plt.title('Scatter plot of Temperature, pH, and Humidity against Plant Names')    # name of Chart
plt.legend()  # printing legend on Graph
plt.show()    # plotting actual Graph
```

```python
# Dividing X and Y data in 80:20 ratio
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42) # (test size is 0.2 means 20)
```

```python
# Create a Random Forest Classifier with 100 estimators
clf = RandomForestClassifier(n_estimators=100, random_state=42)
```

**Random Forest Classifier :** Random forest is a commonly-used machine learning algorithm which combines the output of multiple decision trees to reach a single result. It handles both classification and regression problems.

**n_estimators :** This is the number of trees you want to build before taking the maximum voting or averages of predictions. Higher number of trees give you better performance but makes your code slower. By default 100 is use.

**random_state :** It is a parameter in train_test_split that controls the random number generator used to shuffle the data before splitting it. In other words, it ensures that the same randomization is used each time you run the code, resulting in the same splits of the data. 42 is most commonly used value.

```python
#fitting trained data
clf.fit(X_train, y_train)
```

```python
# using x-test data for prediction so that it can be used for comparision with y-test
y_pred = clf.predict(X_test)
```

```python
# creating classfication report based on y-test (labels) abd y-pred (predicted result based on x-test)
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
# Printing Accuracy
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
#prediction for new values
new_temperature = 25
new_ph = 7
new_humidity = 54
```

```python
# Making prediction for the new values
new_prediction = clf.predict([[new_temperature, new_ph, new_humidity]])
```
